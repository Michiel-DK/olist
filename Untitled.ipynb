```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from stt.profitability import Profitability
```

```python
optim_df, to_cut = Profitability().optim_df()
```

```python
from stt.items import Items
```

```python
itemsprod = Items().product_cat()
```

```python
itemsproducts = itemsprod
```

```python
itemsproducts = pd.merge(itemsproducts, to_cut, on='seller_id', how="left")
itemsproducts['cut'] = itemsproducts['cut'].apply(lambda x: 'cut' if x == 'cut' else 'keep')
```

```python
itemsproducts
```

```python
itemprod_pivot = pd.pivot_table(itemsproducts, columns = 'cut', index = 'product_category_name', values = 'order_item_id', aggfunc = np.sum,fill_value = 0)
```

```python
itemprod_pivot
```

```python
itemprod_pivot.sort_values(by='cut',ascending=False)
```

```python
itemprod_pivot['total'] = itemprod_pivot['keep'] + itemprod_pivot['cut']
itemprod_pivot['%pcs_cut'] = itemprod_pivot['cut'] / itemprod_pivot['total']
itemprod_pivot['50%_treshold'] = itemprod_pivot['%pcs_cut'].apply(lambda x: ">50%" if x > 0.5 else "<50%")
```

```python
itemprod_pivot
```

```python
itemprod_pivot[itemprod_pivot['50%_treshold'] == '>50%']
```

```python
# % of product categories affected
len(itemprod_pivot[itemprod_pivot['50%_treshold'] == '>50%'])/len(itemprod_pivot)
```

```python
itemprod_pivot[itemprod_pivot['50%_treshold'] == '<50%']
```

```python
itemprod_pivot
```

```python
top_orders = Items().get_top_orders()
top_volume = Items().get_top_volume()
```

```python
top_orders
```

```python
top_volume
```

```python
cut_prod = itemprod_pivot.merge(top_orders, on='product_category_name').merge(top_volume, on='product_category_name')
```

```python
cut_prod
```

```python
sns.relplot(x="%pcs_cut", y=cut_prod.index, hue="50%_treshold", size="# items ordered",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=12, data=cut_prod.sort_values(by=["50%_treshold",'# items ordered'], ascending=False))
```

```python
sns.relplot(x="%pcs_cut", y=cut_prod.index, hue="50%_treshold", size="# volume ordered",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=10, data=cut_prod.sort_values(by='# volume ordered', ascending=False))
```

```python
from stt.data import Olist
```

```python
data = Olist().get_data()
```

```python
seller = data['sellers']
geolocation = data['geolocation']
```

```python
geolocation.rename(columns={'geolocation_zip_code_prefix':'seller_zip_code_prefix'},inplace=True)
```

```python
geolocation
```

```python
zipcode = geolocation.groupby('seller_zip_code_prefix').agg({'geolocation_lat':'mean','geolocation_lng':'mean'})
```

```python
zipcode
```

```python
seller_geo = seller.merge(zipcode, on='seller_zip_code_prefix')
```

```python
seller_cut = items_geo[['seller_id','cut','seller_city','seller_state']].drop_duplicates()
```

```python
import plotly.express as px
```

```python
mp = items_geo.groupby('seller_id').agg({'order_id':'nunique','geolocation_lat':'mean','geolocation_lng':'mean'}).merge(seller_cut, on='seller_id')
```

```python
mp
```

```python
fig = px.scatter_geo(mp[mp['cut']=='cut'], lat="geolocation_lat", lon="geolocation_lng", 
                    size="order_id", title="Geographic Spread Cut Sellers",
                     scope = 'south america',
                     opacity = 0.5,
                     center = {'lat':-19.8, 'lon':-43})
fig.update_geos(fitbounds="locations")
fig.show()
```

```python
fig = px.scatter_geo(mp[mp['cut']!='cut'], lat="geolocation_lat", lon="geolocation_lng", 
                    size="order_id", title="Geographic Spread Olist",
                     scope = 'south america',
                     opacity = 0.5,
                     center = {'lat':-19.8, 'lon':-43})
fig.update_geos(fitbounds="locations")
fig.show()
```

```python
state_pivot = pd.pivot_table(mp, columns = 'cut', index = 'seller_state', values = 'seller_id', aggfunc = 'count',fill_value = 0)
```

```python
state_pivot
```

```python
plt.figure(figsize=(20,6))
plt.bar(state_pivot.index, state_pivot['keep'], label='Sellers to keep', color='g')
plt.bar(state_pivot.index, state_pivot['cut'], bottom=state_pivot['keep'], label='Sellers to cut', color='r')
plt.legend(fontsize=15)
```
