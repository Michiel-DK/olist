```python
from stt.profitability import Profitability
```

```python
seller_rc = Profitability().merge_table()
```

```python
seller_rc.columns
```

```python
summary = seller_rc.groupby('profitable').agg({'seller_id':'count','n_orders':'sum','cost_reviews':'sum','profit':'sum','review_score': 'mean','wait_time':'mean'})
```

```python
import matplotlib.pyplot as plt
summary
```

```python
f, (ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
ax1.set_title('Split profitability per unique seller')
ax1.pie(summary['seller_id'], explode=(0, 0.2), autopct=f'%1.1f%%', labels= ['non-profitable', 'profitable'],
        shadow=True, startangle=120, textprops=dict(color="black",fontsize=10))
ax1.axis('equal')
ax2.set_title('Split profitability per total amount of orders')
ax2.pie(summary['n_orders'], explode=(0, 0.15), autopct='%1.1f%%',labels= ['non-profitable', 'profitable'],
        shadow=True, startangle=120, textprops=dict(color="black",fontsize=10))
ax2.axis('equal')
```
