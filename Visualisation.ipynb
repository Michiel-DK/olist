## Profitability

>> Profitability calculated based on below assumptions: <br>
>>> Revenue <br>
>>>> -> Olist takes a 10% cut on the product price (excl. freight) of each order delivered. <br>
>>>> -> Olist charges 80 BRL by month per seller.<br>

>>> Cost <br>
>>>> -> 1, 2 and 3 stars review - respectively 100, 50 and 40 BRL incurred <br>
>>>> -> total cumulated IT Costs to be proportional to the square-root of the total cumulated number of orders processed. Since the birth of the marketplace, cumulated IT costs have amounted to 500,000 BRL.

```python
Viz().profit_viz()
```

```python
Viz().cumul_viz(optimisation = 'no')
```

>> Majority of costs due to sellers with bad reviews <br>
>> profit around 500k

```python
Viz().cumul_viz(optimisation = 'yes')
```

>> Optimisation loop created to maximize profit while cutting sellers starting from least profitable <br>
>> Profit increased to around 900K
