```python
import pandas as pd
from scipy.spatial.distance import euclidean, cityblock
import numpy as np
```

```python
coord_count = df.groupby('order_id')[['order_item_id']].count()
more1 = coord_count[coord_count['order_item_id']>1]
```

```python
df['xy'] = df['xy'].map(eval)
df['xy'] = df['xy'].map(list)
```

```python
%%time
# Loop to assign distance0 in 'json format'
dic = []
tracker = 0
for i in more1.index:
    dic.append({
        'id' : i,
        'product_id' : [i for i in df[df['order_id']==i]['product_id']],
        'sku' : [i for i in df[df['order_id']==i]['order_item_id']],
        'xy' : [[0,0]] + ([i for i in df[df['order_id']==i]['xy']]),
        'distance0' : [i for i in df[df['order_id']==i]['distance0']]
        })
    tracker += 1
    print(tracker)
```

```python
%%time
#Adding distance steps
for i in range(len(dic)):
    if len(dic[i]['xy']) == 1:
        dic[i]['distance_steps'] = dic[i]['distance0']
    else:
        for j in dic[i]['xy']:
            dic[i]['distance_steps'] = [cityblock(dic[i]['xy'][j],dic[i]['xy'][j+1]) for j in range(len(dic[i]['xy'])-1)]
```

```python
%%time
#Adding total distance
for i in range(len(dic)):
    dic[i]['distance_total'] = sum(dic[i]['distance_steps'])
```
